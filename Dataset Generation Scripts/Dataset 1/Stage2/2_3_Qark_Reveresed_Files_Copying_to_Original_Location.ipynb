```python
import os
import shutil
```

```python
COPY_FROM_PATH = "I:/Qark_R8/R8_S1"
COPY_TO_PATH = "I:/PhD/Model_Building/APKs_and_GitRepos/8_Round"

no_analysis =[]

files = os.listdir(COPY_FROM_PATH)
count =1
for file in files:
    try:
        print (str(count)+": Copying file :" + file)
        src_path = COPY_FROM_PATH+"/"+file+"/Qark_Reversed"
        print("src_path: ", src_path)
        os.chdir(COPY_TO_PATH+"/"+file)
        
        if os.path.exists('Qark_Reversed'):
            shutil.rmtree("Qark_Reversed")            
        
        dst_path=COPY_TO_PATH+"/"+file
        #shutil.copy(src_path, dst_path)
        shutil.move(src_path, dst_path)
        print("dst_path: ", dst_path)       
        
        count = count+1
        
    except Exception as e:
        print(e)
        no_analysis.append(file)
        continue
print (no_analysis)
```
