```python
import shutil
import os
import zipfile
```

```python
DOWNLOAD_BASE_PATH = "D:/PhD/Model_Building/APKs_and_GitRepos/4_Round"
```

```python
def create_zip_file(path, file_name):
    print("Creating zip file from original source code")
    path = path+"/"+file_name+"/app"    
    zipfile_handler = zipfile.ZipFile(file_name+'.zip', 'w', zipfile.ZIP_DEFLATED)        
    
    for root, dirs, files in os.walk(path):
        for file in files:
            zipfile_handler.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), os.path.join(path, '..')))
            
    zipfile_handler.close()
```

```python
def rename_git_repo_folder(path, current_name, new_name):
    print ("Renaming current GIT repo folder name : "+current_name+" to :"+new_name)
    os.chdir(path)    
    os.rename(current_name, new_name)
```

```python
def execute_process():
    print("***************Process started***************")
    files = os.listdir(DOWNLOAD_BASE_PATH)
    
    for file in files:
        print("***************App Name : "+file+"***************")
        sub_files = os.listdir(DOWNLOAD_BASE_PATH+"/"+file)        
        for sub_file in sub_files:      
            
            try:
                if(str(sub_file) not in [file+".apk", "MobSF_Reversed", "Qark_Reversed", file+".zip"]):
                    path = DOWNLOAD_BASE_PATH+"/"+file
                    rename_git_repo_folder(path ,sub_file, file)
                    create_zip_file(path, file)
            except Exception as e:
                print(e)
    print("***************Process completed***************")
```

```python
execute_process()
```
