```python
MobSF_vulnerability_df_1_round_Files = pd.read_csv(MobSF_csv_1_round_Files, low_memory=False).fillna("")
MobSF_vulnerability_df_2_round_Files = pd.read_csv(MobSF_csv_2_round_Files, low_memory=False).fillna("")
MobSF_vulnerability_df_3_round_Files = pd.read_csv(MobSF_csv_3_round_Files, low_memory=False).fillna("")
MobSF_vulnerability_df_4_round_Files = pd.read_csv(MobSF_csv_4_round_Files, low_memory=False).fillna("")
MobSF_vulnerability_df_5_round_Files = pd.read_csv(MobSF_csv_5_round_Files, low_memory=False).fillna("")
MobSF_vulnerability_df_6_round_Files = pd.read_csv(MobSF_csv_6_round_Files, low_memory=False).fillna("")
MobSF_vulnerability_df_7_round_Files = pd.read_csv(MobSF_csv_7_round_Files, low_memory=False).fillna("")

vulnerability_df_MobSF= pd.concat([MobSF_vulnerability_df_1_round_Files, MobSF_vulnerability_df_2_round_Files, MobSF_vulnerability_df_3_round_Files, MobSF_vulnerability_df_4_round_Files, MobSF_vulnerability_df_5_round_Files, MobSF_vulnerability_df_6_round_Files, MobSF_vulnerability_df_7_round_Files], ignore_index=True)
vulnerability_df_MobSF=vulnerability_df_MobSF.drop_duplicates(subset=['Code', 'CWE_ID'], keep='first')


Qark_vulnerability_df_1_round_Files = pd.read_csv(Qark_csv_1_round_Files, low_memory=False).fillna("")
Qark_vulnerability_df_2_round_Files = pd.read_csv(Qark_csv_2_round_Files, low_memory=False).fillna("")
Qark_vulnerability_df_3_round_Files = pd.read_csv(Qark_csv_3_round_Files, low_memory=False).fillna("")
Qark_vulnerability_df_4_round_Files = pd.read_csv(Qark_csv_4_round_Files, low_memory=False).fillna("")
Qark_vulnerability_df_5_round_Files = pd.read_csv(Qark_csv_5_round_Files, low_memory=False).fillna("")
Qark_vulnerability_df_6_round_Files = pd.read_csv(Qark_csv_6_round_Files, low_memory=False).fillna("")
Qark_vulnerability_df_7_round_Files = pd.read_csv(Qark_csv_7_round_Files, low_memory=False).fillna("")

vulnerability_df_Qark= pd.concat([Qark_vulnerability_df_1_round_Files, Qark_vulnerability_df_2_round_Files, Qark_vulnerability_df_3_round_Files, Qark_vulnerability_df_4_round_Files, Qark_vulnerability_df_5_round_Files, Qark_vulnerability_df_6_round_Files, Qark_vulnerability_df_7_round_Files], ignore_index=True)
vulnerability_df_Qark=vulnerability_df_Qark.drop_duplicates(subset=['Code', 'CWE_ID'], keep='first')
```
