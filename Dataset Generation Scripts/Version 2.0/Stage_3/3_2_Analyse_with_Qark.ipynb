```python
import subprocess
import os
import json
import csv
from datetime import datetime
```

```python
DOWNLOAD_BASE_PATH = "I:/PhD/Model_Building/APKs_and_GitRepos/GITRepos_Src_Only/7_Round"
QARK_BASE_PATH = "I:/PhD/Model_Building/Scanners_and_Decompilers/Scanners/qark"
REPORT_PATH = "C:/Users/janak/AppData/Local/Programs/Python/Python39/Lib/site-packages/qark/report"
#APPLICATION_BASE_PATH = "D:/PhD/Model_Building/Stage_2_Analysing_APKs"
APPLICATION_BASE_PATH = "I:/PhD/Model_Building/APKs_and_GitRepos/GITRepos_Src_Only/7_Round"
```

```python
def start_scan(app_name):
    os.chdir(QARK_BASE_PATH)
#     file_param = DOWNLOAD_BASE_PATH+"/"+app_name+"/"+app_name
    file_param = DOWNLOAD_BASE_PATH+"/"+app_name
    command = "qark_execute_files.bat "+file_param
    subprocess.call(command, creationflags=subprocess.CREATE_NEW_CONSOLE) 
```

```python
def write_to_csv(data_row):
    scanned_App_Details_file = open(APPLICATION_BASE_PATH+'/Qark_Scanned_File_Details.csv', 'a', newline='')
    scanned_App_Details_file_writer = csv.writer(scanned_App_Details_file)       
    scanned_App_Details_file_writer.writerow(data_row)  
    scanned_App_Details_file.close()
```

```python
def get_report_details(app_name):
    report_file = open(REPORT_PATH+"/report.json")
    report_data = json.load(report_file)
    
    #header_csv_row = ['app_name','category','severity','description','name','line_number','file_object','apk_exploit_dict']
    #write_to_csv(header_csv_row);
 
    for issue in report_data:
        app_name = app_name 
        issue_category = issue['category']
        issue_severity = issue['severity']
        issue_description = issue['description']
        issue_name = issue['name']
        issue_line_number = str(issue['line_number']).replace(",", "_")
        issue_file_object = issue['file_object']
        issue_apk_exploit_dict = str(issue['apk_exploit_dict']).replace(",", "_")
        issue_csv_row= [app_name, issue_category,issue_severity,issue_description,issue_name,issue_line_number,issue_file_object,issue_apk_exploit_dict]
        write_to_csv(issue_csv_row);
 

    report_file.close()
```

```python
def execute_process():
    print("***************Process started***************")
    files = os.listdir(DOWNLOAD_BASE_PATH)
    for file in files:  
        try:
            print(str(datetime.now())+" :***************File Name : "+file+"***************")        
            app_name = file
            print("***************Scanning APK***************")
            start_scan(app_name)
            print("***************APK Scanned***************")
            print("***************Generating Report***************")
            get_report_details(app_name)
            print("***************Report Generated***************")
        except Exception as e:
            print(e)
            continue
        
    print("***************Process completed***************")
```

```python
execute_process()
```
