```python
import os
import shutil
```

```python
COPY_FROM_PATH = "D:/PhD/Model_Building/APKs_and_GitRepos/2_Round"
COPY_TO_PATH = "D:/PhD/Model_Building/APKs_and_GitRepos/2_Round_New"
```

```python
def move_apks():
    os.chdir(COPY_FROM_PATH) 
    files = os.listdir(COPY_FROM_PATH)
    count =1

    os.chdir(COPY_TO_PATH) 
    for file in files:

        try:
            print (str(count)+": Copying file :" + file)

            try:

                APK_src_path = COPY_FROM_PATH+"/"+file+"/"+file+".apk"
                APK_dst_path = COPY_TO_PATH+"/"+"APKs"

                if os.path.exists(APK_dst_path+"/"+file):
                    shutil.rmtree(APK_dst_path+"/"+file)    

                os.chdir(APK_dst_path)
                os.mkdir(file)

                APK_dst_path = APK_dst_path+"/"+file            
                shutil.move(APK_src_path, APK_dst_path)            
                print("Copying from : "+APK_src_path)
                print("Copied to : "+APK_dst_path)
            except Exception as e:
                print(e)
                print("APK Copying unsuccessful!")
                continue
            finally:
                os.chdir(COPY_FROM_PATH) 
                print(os.getcwd())                 
            count = count+1


        except Exception as e:
            print(e)

            continue

        finally:
            os.chdir(COPY_FROM_PATH) 
```

```python
def move_repos():
    os.chdir(COPY_FROM_PATH) 
    files = os.listdir(COPY_FROM_PATH)
    count =1

    os.chdir(COPY_TO_PATH) 
    for file in files:

        try:
            print (str(count)+": Copying GIT Repo :" + file)       


            try:
                GIThub_Repo_src_path = COPY_FROM_PATH+"/"+file+"/"+file
                GIThub_Repo_dst_path = COPY_TO_PATH+"/"+"GITRepos"

                if os.path.exists(GIThub_Repo_dst_path+"/"+file):
                    shutil.rmtree(GIThub_Repo_dst_path+"/"+file)    

                os.chdir(GIThub_Repo_dst_path)

                shutil.move(GIThub_Repo_src_path, GIThub_Repo_dst_path)
                print("Copying from : "+GIThub_Repo_src_path)
                print("Copied to : "+GIThub_Repo_dst_path)

            except Exception as e:
                print(e)
                print("GIThub_Repo Copying unsuccessful!")
                continue
            finally:
                os.chdir(COPY_FROM_PATH) 
                print(os.getcwd())


            count = count+1

        except Exception as e:
            print(e)

            continue

        finally:
            os.chdir(COPY_FROM_PATH) 
```

```python
def mobsf_source_zip():
    os.chdir(COPY_FROM_PATH) 
    files = os.listdir(COPY_FROM_PATH)
    count =1

    os.chdir(COPY_TO_PATH) 
    for file in files:

        try:
            print (str(count)+": Copying Zip file for MobSF_Source :" + file)

            try:

                GIThub_Source_Zip_src_path = COPY_FROM_PATH+"/"+file+"/"+file+".zip"
                GIThub_Source_Zip_dst_path = COPY_TO_PATH+"/"+"GIT_Source_Zip"

                if os.path.exists(GIThub_Source_Zip_dst_path+"/"+file):
                    shutil.rmtree(GIThub_Source_Zip_dst_path+"/"+file)    

                os.chdir(GIThub_Source_Zip_dst_path)
                os.mkdir(file)

                GIThub_Source_Zip_dst_path = GIThub_Source_Zip_dst_path+"/"+file            
                shutil.move(GIThub_Source_Zip_src_path, GIThub_Source_Zip_dst_path)            
                print("Copying from : "+GIThub_Source_Zip_src_path)
                print("Copied to : "+GIThub_Source_Zip_dst_path)
            except Exception as e:
                print(e)
                print("Zip File Copying unsuccessful!")
                continue
            finally:
                os.chdir(COPY_FROM_PATH) 
                print(os.getcwd())         



            count = count+1

        except Exception as e:
            print(e)

            continue

        finally:
            os.chdir(COPY_FROM_PATH) 
```

```python
def execute():
    print("#############APK##########")
    move_apks()
    print()
    print("###########Repos##########")
    move_repos()
    print()
    print("#############RepoZip##########")
    mobsf_source_zip()
    print()
    print("#############MobSFReversed##########")
    move_mobsf_reversed()
    print()
    print("#############QarkReveresed##########")
    move_qark_reversed()
```
