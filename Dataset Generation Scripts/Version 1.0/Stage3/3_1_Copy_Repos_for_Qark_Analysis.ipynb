```python
import os
import shutil
```

```python
DOWNLOAD_BASE_PATH = "D:/PhD/Model_Building/APKs_and_GitRepos/2_Round/GITRepos"
COPY_TO_PATH = "D:/PhD/Model_Building/APKs_and_GitRepos/2_Round/GitRepos_Src"

files = os.listdir(DOWNLOAD_BASE_PATH)

for file in files:
    try:
        src_path = DOWNLOAD_BASE_PATH+"/"+file+"/"+file+"/"+"app"
        os.chdir(COPY_TO_PATH)
        os.mkdir(file)
        dst_path=COPY_TO_PATH+"/"+file+"/"+file
        shutil.copytree(src_path, dst_path)
    except Exception as e:
        print(e)
        continue
```
